# Spin-boson dephasing
Assuming an Hamiltonian
$$ H = \frac12\omega_0\sigma_z + \sum_k\omega_k b^\dagger_kb_k + \frac12\sigma_z\otimes \sum_k(g_kb^\dagger_ke^{i \omega_k t}+g_k^*b_ke^{-i \omega_k t}). $$
For initial states
$$ \rho_{SE}(0) = \rho_S\otimes\rho_{th}^{\beta},\qquad\rho_{th}^{\beta} = \frac{e^{-\beta H_E}}{Z_\beta} $$
the reduced dynamics is generated by the GKSL ME
$$ \mathcal L_t[\rho] = \frac d{dt}\rho = \gamma(t)(\sigma_z\rho\sigma_z-\rho) $$
where
$$ \gamma(t) = \sum_k\lvert g_k\rvert^2 \coth\left(\frac{\beta\omega_k}2\right)\frac{\sin\omega_kt}{\omega_k}. $$

## Zero discord initially correlated system-environment
Let us assume that the initial state is correlated but with zero discord
$$ \rho_{SE}(0) = p \lvert0\rangle\langle0\rvert\otimes\rho_{th}^{\beta_0} + (1-p)\lvert1\rangle\langle1\rvert\otimes\rho_{th}^{\beta_1} $$
and let
$$ \mathcal L^i_t[\rho] = \frac d{dt}\operatorname{tr}_E[U(t)(\rho\otimes\rho_{th}^{\beta_i})U^\dagger(t)] = \gamma_i(t)(\sigma_z\rho\sigma_z-\rho) $$
be the generators that one obtains using $\rho_{th}^{\beta_i}$ as initial state.

### B+ decomposition
The B+ decomposition for a zero discord initial state reads
$$ w_0Q_0 = w_xQ_x = w_yQ_y = p\rho_{th}^{\beta_0} + (1-p)\rho_{th}^{\beta_1},\qquad w_z\rho_z = 2(1-p)\rho_{th}^{\beta_1}. $$
Let 
$$ \tilde{\mathcal L}^\alpha_t[\rho] = \frac d{dt}\operatorname{tr}_E[U(t)(\rho\otimes\rho_\alpha)U^\dagger(t)] $$
be the generators corresponding to the initial states $\rho_\alpha$ of the B+ decomposition.
These new generators will be again dephasing with rates
$$ \tilde\gamma_{0,x,y}(t) = p \gamma_0(t) + (1-p)\gamma_1(t),\qquad\tilde\gamma_z(t) = \gamma_1(t). $$

### System-side repreparations
Let
$$ \rho_{SE}^{\mathcal R}(0) = (\mathcal R\otimes\operatorname{id})\rho_{SE}(0) $$
be a system-only repreparation of the initial state.
Let $\mathcal R(Q_\alpha) = \sum_{\alpha,\alpha^\prime}R_{\alpha,\alpha^\prime}Q_{\alpha^\prime}$.
We can obtain the time evolution of these initial states as
$$ \rho_{SE}^{\mathcal R}(t) = \frac{\sum_{\alpha,\alpha^\prime}w_\alpha R_{\alpha,\alpha^\prime}\Phi^\alpha_t(Q_{\alpha^\prime})}{\sum_\alpha w_\alpha\operatorname{tr}[\mathcal R(Q_\alpha)]}. $$
For the example of the dephasing, we only need to calculate three extra unravelings $\Phi^z_t(Q_{0,x,y})$ to be able to calculate all initial states $\rho_{SE}^{\mathcal R}$ obtained from repreparations.

#### Examples
As a first initial state let us consider
$$ \rho_{SE}^{\mathcal R_f} = p \lvert1\rangle\langle1\rvert\otimes\rho_{th}^{\beta_0} + (1-p)\lvert0\rangle\langle0\rvert\otimes\rho_{th}^{\beta_1} $$
obtained with a bit-flip, i.e. a the repreparation $\mathcal R_f(\rho) = \sigma_x\rho\sigma_x$ such that
$$ \mathcal R_f(\lvert0\rangle\langle0\rvert) = \lvert1\rangle\langle1\rvert,\qquad \mathcal R_f(\lvert1\rangle\langle1\rvert) = \lvert0\rangle\langle0\rvert. $$
In other words, we are simply swapping the environmental states based on the initial system state.
If we write $\mathcal R_f(Q_\alpha) = \sum_{\alpha^\prime}R^f_{\alpha,\alpha^\prime}Q_\alpha^\prime$, the non-zero elements of the matrix R^1 are
$$ R^f_{00}=R^f_{xx}=1,\quad R^f_{yy}R^f_{zz}=-1,\quad R^f_{0y}=R^f_{0z}=2. $$

Similarly, we can consider a repreparation $\mathcal R_H(\rho) = H\rho H$, where $H$ is the Hadamard operator $H\lvert0\rangle = \lvert-\rangle$, $H\lvert1\rangle = \lvert+\rangle$, so that 
$$ \rho_{SE}^{\mathcal R_H} = p \lvert-\rangle\langle-\rvert\otimes\rho_{th}^{\beta_0} + (1-p)\lvert+\rangle\langle+\rvert\otimes\rho_{th}^{\beta_1}. $$
The non-zero elements of the matrix $R^H$ are
$$ R^H_{00} = R^H_{xz}=R^H_{zx}=1,\quad R^H_{0y}=2,\quad R^H_{yy}=-1. $$

Since any repreparation $\mathcal R$ (not necessarely unitary) can be applied, from the 6 unravelings calculated we can get the dynamics of any initial system state with purity less or equal to $\rho_{SE}(0)$.
